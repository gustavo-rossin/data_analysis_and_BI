```python
import xlsxwriter as xw
import os
import random
import pyautogui as pag
import numpy as np
```

```python
nome_arquivo = 'exemplo1.xlsx'

wb = xw.Workbook(nome_arquivo, options={'strings_to_numbers': True})

s1 = wb.add_worksheet('Planilha_1')

bg_colorido = wb.add_format({'bg_color': 'red', 'font_color': 'white'})
fg_colorido = wb.add_format({'fg_color': 'silver', 'font_color': 'red'})
bold = wb.add_format({'bold': True})

s1.write('A1', 'Nome', bg_colorido)
s1.write('B1', 'Idade', fg_colorido)
s1.set_column('A:B', 24)


random_int = random.randint(3, 10) + 1
print(random_int)

for nome in range(2, random_int):
    s1.write('A' + str(nome), f'Nome{nome}')
    s1.write('B' + str(nome), f'{int(random.randint(18, 65))}')
s1.write('A' + str(random_int), 'Total', bold)
s1.write_formula('B' + str(random_int), f'=SUM(B2:B{random_int - 1})')
pag.sleep(2)
wb.close()

os.startfile(nome_arquivo)
```

```python
nome_arquivo = 'exemplo_formatacao_condicional.xlsx'

wb = xw.Workbook(nome_arquivo, options={'strings_to_numbers': True})

s1 = wb.add_worksheet('Planilha_1')

bg_colorido = wb.add_format({'bg_color': 'red', 'font_color': 'white'})
fg_colorido = wb.add_format({'fg_color': 'silver', 'font_color': 'red'})
bold = wb.add_format({'bold': True})
merge_cells = wb.add_format({
    'bold': True,
    'border': 6,
    'valign': 'vcenter',
    'align': 'center',
    'size': 24,
    'fg_color': 'silver',
    'font_color': 'red',
})

formato_positivo = wb.add_format({
    'bg_color': 'green',
    'font_color': 'white',
})

formato_negativo = wb.add_format({
    'bg_color': 'red',
    'font_color': 'black',
})
data = [41, 50, 29, 37, 81, 30, 73, 63, 20, 35, 68, 22, 60, 31, 95]
inserir_dados = [
    ['Coluna1', 'Coluna2', 'Coluna3', 'Coluna4'],
    [random.choice(data), random.choice(data), random.choice(data), random.choice(data)],
    [random.choice(data), random.choice(data), random.choice(data), random.choice(data)],
    [random.choice(data), random.choice(data), random.choice(data), random.choice(data)],
    [random.choice(data), random.choice(data), random.choice(data), random.choice(data)],
]

for linha, dados in enumerate(inserir_dados):
    s1.write_row(linha + 4, 1, dados)
    

s1.conditional_format('B6:E9', {'type': 'cell',
                                'criteria': '>=',
                                'value': np.mean(inserir_dados[1:]),
                                'format': formato_positivo
                               })
s1.conditional_format('B6:E9', {'type': 'cell',
                                'criteria': '<',
                                'value': np.mean(inserir_dados[1:]),
                                'format': formato_negativo
                               })

print('media que varia com a escolha aleatoria da data:', np.mean(inserir_dados[1:]))
s1.merge_range('A1:F4', 'Formatação Condicional', merge_cells)

wb.close()

os.startfile(nome_arquivo)
```

```python
# GRAFICO

nome_arquivo = 'exemplo_grafico.xlsx'

wb = xw.Workbook(nome_arquivo, options={'strings_to_numbers': True})

s1 = wb.add_worksheet('Gráfico')

bg_colorido = wb.add_format({'bg_color': 'red', 'font_color': 'white'})
fg_colorido = wb.add_format({'fg_color': 'silver', 'font_color': 'red'})
bold = wb.add_format({'bold': True})
merge_cells = wb.add_format({
    'bold': True,
    'border': 6,
    'valign': 'vcenter',
    'align': 'center',
    'size': 24,
    'fg_color': 'silver',
    'font_color': 'red',
})




data_nome = ['Ana', 'Gustavo', 'Sattyla', 'Andre', 'Vanessa', 'Pietra', 'Guilherme', 'Jojo', 'Yolanda', 'Juan']
data_valores = [120, 938, 182, 666, 291, 901, 400, 201, 89, 766, 455, 119]
inserir_dados = [
    ['vendedores', 'total_vendas'],
    [random.choice(data_nome), random.choice(data_valores)],
    [random.choice(data_nome), random.choice(data_valores)],
    [random.choice(data_nome), random.choice(data_valores)],
    [random.choice(data_nome), random.choice(data_valores)],
]

s1.write('A11', 'Total', bg_colorido)
s1.write_formula('B11', '=SUM(B6:B10)', bg_colorido)
s1.set_column('A:B', 24)

for linha, dados in enumerate(inserir_dados):
    s1.write_row(linha + 5, 0, dados)
    
s1.merge_range('A1:F4', 'Grafico', merge_cells)

grafico_colunas = s1.add_chart({'type': 'column'})

grafico_colunas.add_series({'name': '=Gráfico!$B$1', 'categories': '=Gráfico!$A$7:$A$11'}, 'values': '=Gráfico$B$7:$B$11')

wb.close()

os.startfile(nome_arquivo)
```
