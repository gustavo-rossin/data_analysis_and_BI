```python
"""
Exercicio 1

Desenvolver um programa que solicita ao usuário a entrada de um número e, com base nesse número, realiza as seguintes operações:

1 - mostra o número informado
2- informa se é par ou impar
3 informa se é positivo, negativo ou zero
4 - Se for positivo, calcular e mostrar sua raiz quadrada

"""

def informacao_numero():
    numero = float(input('Digite um número qualquer:\n'))
    print(f'O número digitado foi: {numero}')
    
    if numero % 2 == 0 and numero > 0:
        print(f'O número {numero} é PAR e POSITIVO')
        print(f'Sua raiz quadrada é: {numero ** (1/2):.2f}')
    elif numero % 2 == 0 and numero == 0:
        print(f'O número {numero} é PAR e ZERO')
    elif numero % 2 == 0 and numero < 0:
        print(f'O número {numero} é PAR e NEGATIVO')
    else:
        print(f'O número {numero} é IMPAR')
        if numero < 0:
            print(f'O número {numero} é NEGATIVO')
        else:
            print(f'O número {numero} é POSITIVO')
            print(f'Sua raiz quadrada é: {numero ** (1/2):.2f}')
    
informacao_numero()
```

```python
"""
Exercício 2 
Operações Matemáticas Básicas

Objetivo: Desenvolver um programa que solicite ao usuário a entrada de dois números e, com base nesses números, realize as seguintes operações:

    1. Calcular e mostrar a soma dos dois números.
    2. Calcular e mostrar a subtração do primeiro número pelo segundo.
    3. Calcular e mostrar a multiplicação dos dois números.
    4. Calcular e mostrar a divisão do primeiro número pelo segundo (se o segundo número não for zero).
    5. Verificar e informar se algum dos números (ou ambos) é zero.
    6. Calcular e mostrar a média dos dois números.
    7. Comparar os dois números e informar qual é maior ou se são iguais.
"""

def operacoes_dois_numeros():
    num1 = float(input('Digite o primeiro número:\n'))
    num2 = float(input('Digite o segundo número:\n'))

    soma = num1 + num2
    subtracao = num1 - num2
    multiplicacao = num1 * num2
    media = soma / 2
    comparacao = ['São Iguais' if num1 == num2 else max(list((num1, num2)))]
    try:
        divisao = num1 / num2
        return f"""
        A soma dos números: {num1} + {num2} = {soma}
        A subtração dos números: {num1} - {num2} = {subtracao}
        A multiplicação dos números: {num1} * {num2} = {multiplicacao}
        A divisão dos números: {num1} / {num2} = {divisao}
        A média dos números: {media}
        O maior número entre os dois: {comparacao[0]}"""
    except ZeroDivisionError:
        return 'Não é possível dividir por ZERO'
    
print(operacoes_dois_numeros())
```

```python
"""
Exercício Cálculo 4
Salário com Horas Extras e Imposto de Renda

Objetivo: Desenvolver um programa que ajude o usuário a entender a 
composição do seu salário mensal. O programa deve considerar o valor 
recebido por hora, as horas trabalhadas no mês, o cálculo de horas 
extras e o desconto do imposto de renda.

Instruções:

    1. Solicite ao usuário o valor que ele ganha por hora.
    2. Pergunte quantas horas foram trabalhadas no mês.
    3. Calcule o salário bruto multiplicando o valor por hora 
    pelas horas trabalhadas.
    
    4. Se o usuário tiver trabalhado mais de 160 horas no mês, 
    calcule o valor das horas extras. Cada hora extra deve ser 
    compensada com um adicional de 50% sobre o valor da hora comum.
    
    5. Calcule o imposto de renda sobre o salário (considerando as horas extras). 
    O imposto de renda tem uma taxa fixa de 11% sobre o salário.
    
    6. Mostre ao usuário uma descrição detalhada contendo:
        - Salário Bruto.
        - Valor das Horas Extras (se aplicável).
        - Valor do Imposto de Renda.
        - Salário Líquido (Salário Bruto + Horas Extras - Imposto de Renda).
"""

salario_por_hora = float(input('Quanto você recebe por hora trabalhada?\n'))
horas_trabalhadas_mes = int(input('Quantas horas você trabalha no mês?\n'))
salario_bruto = salario_por_hora * horas_trabalhadas_mes

print(f'Salário Bruto = {salario_bruto}')

if horas_trabalhadas_mes > 160:
    horas_extras = (horas_trabalhadas_mes - 160)
    pagamento_hora_extra = 1.5 * salario_por_hora
    salario_descontado_imposto = salario_bruto * 0.89
    salario_liquido = salario_descontado_imposto + (pagamento_hora_extra * horas_extras)
    
    print(f'Horas Extras: {horas_extras}')
    print(f'Recebimento a mais: R$ {pagamento_hora_extra * horas_extras}')
    print(f'Valor Imposto de Renda: R$ {salario_bruto * 0.11}')
    print(f'Salário Líquido: R$ {salario_liquido}')
    
else:
    print(f'Valor Imposto de Renda: R$ {salario_bruto * 0.11}')
    print(f"Salário Líquido: R$ {salario_bruto * 0.89}")
        
```

```python
"""
Exercício 7
Análise Comparativa e Estatística de Números

Objetivo: Desenvolver um programa que obtenha do usuário dez números e realize 
análises estatísticas e matemáticas com base nos valores inseridos.

Instruções:

    1. Solicite ao usuário que informe dez números.
    2. Determine e exiba qual é o maior número.
    3. Determine e exiba qual é o menor número.
    4. Calcule e exiba a média dos dez números.
    5. Informe quantos dos números são pares.
    6. Informe quantos dos números são positivos.
    7. Calcule e exiba a variação (diferença) entre o maior e o menor número.
    8. Mostre a soma dos números que são maiores que a média.
    9. Informe quantos dos números são negativos.
    
Observações: O programa deve considerar que o usuário fornecerá apenas 
valores numéricos válidos. Em todas as operações, os resultados devem ser 
exibidos com até duas casas decimais, quando aplicável.

"""

def analise_comparativa_e_estatistica():
    contador = 1
    new_lista = []
    is_even = []
    is_positive = []
    sum_mean = []
    is_negative = []
    zero = []
    while contador <= 10:
        num = int(input('Digite um número inteiro: '))
        new_lista.append(num)
        contador += 1
        if num % 2 == 0:
            is_even.append(num)
        if num > 0:
            is_positive.append(num)
    max_number = max(new_lista)
    min_number = min(new_lista)
    dif = max_number - min_number
    media = sum(new_lista) / len(new_lista)
    count_even = len(is_even)
    count_odd = len(new_lista) - count_even
    count_positive = len(is_positive)\

    
    for neg in new_lista:
        if neg < 0:
            is_negative.append(neg)
        elif neg == 0:
            zero.append(neg)
    
    for n in new_lista:
        if n > media:
            sum_mean.append(n)
    
    return f"""
    MAIOR número da lista: {max_number}
    --
    MENOR número da lista: {min_number}
    --
    DIFERENÇA entre MAX e MIN: {dif}
    --
    MEDIA: {media}
    --
    SOMA dos valores ACIMA da MÉDIA: {sum(sum_mean)}
    --
    Contagem de PARES: {count_even}. Lista: {is_even}
    --
    Contagem de ÍMPARES: {count_odd}.
    --
    Contagem de POSITIVOS: {count_positive}. Lista: {is_positive}
    --
    Contagem de NEGATIVOS: {len(is_negative)}. Lista: {is_negative}
    --
    Contagem de ZEROS: {len(zero)}. Lista: {zero}
    
    """



print(analise_comparativa_e_estatistica())
```

```python
"""
Exercício 8
Análise de Caracteres e Formação de Palavras

Objetivo: Desenvolver um programa que obtenha do usuário uma letra ou uma palavra 
e realize análises sobre o tipo de caractere inserido, além de funções adicionais 
se uma palavra completa for fornecida.

Instruções:

    1. Solicite ao usuário que digite uma letra ou uma palavra.
    
    2. Se o usuário digitar apenas um caractere:
    
        a. Verifique se é uma vogal ou consoante.
        b. Se for uma vogal, exiba: "Você digitou uma vogal."
        c. Se for uma consoante, exiba: "Você digitou uma consoante."
        d. Se o caractere não for uma letra, exiba: "Caractere inválido."
    
    3. Se o usuário digitar mais de um caractere (uma palavra):
        
        a. Calcule e exiba o número de vogais na palavra.
        b. Calcule e exiba o número de consoantes na palavra.
        c. Exiba a palavra em ordem inversa.
        d. Informe se a palavra é um palíndromo (se lê da mesma forma de frente para trás 
        e de trás 
        para frente, como "arara").
    
    4. Peça ao usuário se deseja verificar outra letra ou palavra ou sair do programa.

Observações: O programa deve considerar letras maiúsculas e 
minúsculas (ou seja, "A" e "a" são ambas vogais).
"""

def checagem_de_palavras():
    lista_vogais = ['a', 'e', 'i', 'o', 'u']
    palavra = input('Digite uma palavra ou um caractere: \n').lower()
    
    if len(palavra) == 1:
        if palavra in 'aeiou':
            return 'Você digitou uma vogal.'
        elif palavra in 'bcdfghjklmnpqrstuvxwyz':
            return 'Você digitou uma consoante'
        else:
            return 'Caractere inválido'
    else:
        contagem_vogal = 0
        contagem_consoantes = 0
        lista_vogais_palavra = []
        lista_consoantes = []
        lista_palavra = list(palavra)
        palindromo = [True if palavra == palavra[::-1] else False]
        print(list(palavra))
        for letra in lista_palavra:
            if letra in lista_vogais:
                contagem_vogal += 1
                lista_vogais_palavra.append(letra)
                
        for consoante in lista_palavra:
            if consoante not in lista_vogais:
                contagem_consoantes += 1
                lista_consoantes.append(consoante)
            
        
        return f"""
        A palavra {palavra} possui:
        {contagem_vogal} vogal(is). Vogais na palavra: {lista_vogais_palavra}
        {contagem_consoantes} consoante(s). Consoantes na palavra: {lista_consoantes}
        A palavra {palavra} é Palíndromo? {palindromo}
        
        """
            
            
print(checagem_de_palavras())
```

```python
"""
Exercício 28
Contagem Regressiva para Lançamento de Foguete

Objetivo:

Desenvolver um programa em Python que realize uma contagem regressiva 
para o lançamento de um foguete.

Requisitos:

    1. O programa deve iniciar a contagem em 10 e terminar em 0.

    2. Cada número deve ser impresso na tela, um por vez.

    3. Entre cada número impresso, o programa deve fazer uma pausa de 1 segundo.

    4. Após chegar a 0, o programa deve imprimir a palavra "Fogo!" na tela.

Dicas:

    Utilize a função sleep do módulo time para fazer a pausa entre os números.
"""

import time

def decolagem():
    num = int(input('Tempo (em segundos) para a decolagem: '))
    
    for i in range (num, -1, -1):
        print(i)
        time.sleep(1)
        

    print('FIRE')

    
decolagem()
```
