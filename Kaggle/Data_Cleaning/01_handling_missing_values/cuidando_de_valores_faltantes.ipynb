# Explicação

```python
import pandas as pd
import numpy as np

#importando a tabela
melbourne_path = '../../Intro_Machine_Learning/02_basic_data_exploration/melb_data.csv'
melbourne_data = pd.read_csv(melbourne_path)
np.random.seed(0)

melbourne_data.head()
```

```python
# get the number of missing data points per column
count_null = melbourne_data.isnull().sum()

# print(count_null)

features_with_null = ['Car', 'BuildingArea', 'YearBuilt', 'CouncilArea']

melbourne_columns_with_null = melbourne_data[features_with_null].isnull().sum()

print('colunas com null:\n', melbourne_columns_with_null)
```

```python
# how many total missing values do we have?
# (linhas, colunas)
print('(linhas, colunas)', melbourne_data.shape)

total_cells = np.product(melbourne_data.shape)
print('todas as células:', total_cells)

missing_cells = melbourne_columns_with_null.sum()
print('total de nulls:', missing_cells)

percent_missing = (missing_cells/total_cells) * 100
print(format(percent_missing, ".2f"))
```

```python
# remove all the rows that contain a missing value
melbourne_dropna = melbourne_data.dropna()
melbourne_dropna.head()

# remove all columns with at least one missing value
melbourne_columns_na = melbourne_data.dropna(axis=1)

melbourne_columns_na.head()
```

```python
# get a small subset of the NFL dataset
melbourne_subset = melbourne_data[features_with_null]

print('sem colocar fillna: \n', melbourne_subset.head())

# replace all NA's with 0

fillna_melbourne = melbourne_subset.fillna(0)
print('com fillna: \n', fillna_melbourne.head())

# replace all NA's the value that comes directly after it in the same column, 
# then replace all the remaining na's with 0
bfill_melbourne = melbourne_subset.fillna(method='bfill', axis=0).fillna(0)
print('com fillna e bfill:\n ', bfill_melbourne.head())
```
